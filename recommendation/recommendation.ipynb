```python
import pandas as pd

# Load your Excel file (adjust the path if needed)
df = pd.read_excel("C:\\Users\\HFCS\\Desktop\\Project\\recommendation\\Books_details.xlsx", engine='openpyxl')

# Clean column names
df.columns = df.columns.str.strip()

# Convert 'Rate' column to numeric
df['Rate'] = pd.to_numeric(df['Rate'], errors='coerce')

# Drop rows with missing or invalid ratings
df = df.dropna(subset=['Rate'])

# Sort by rating (descending)
top_rated_books = df.sort_values(by='Rate', ascending=False)

# Select top N books
def recommend_top_books(top_n=5):
    return top_rated_books[["Name", "Author", "Description", "Rate", "Number of pages", "thumbnail"]].head(top_n)

# Example usage
recommended_books = recommend_top_books(5)
print(recommended_books)
```

```python
import pandas as pd
import tkinter as tk
from tkinter import ttk
from PIL import Image, ImageTk
import requests
from io import BytesIO

# Load and process the Excel data
df = pd.read_excel("C:\\Users\\HFCS\\Desktop\\Project\\recommendation\\Books_details.xlsx", engine='openpyxl')
df.columns = df.columns.str.strip()
df['Rate'] = pd.to_numeric(df['Rate'], errors='coerce')
df = df.dropna(subset=['Rate'])
top_books = df.sort_values(by='Rate', ascending=False).head(5)

# Create the main window
root = tk.Tk()
root.title("أفضل 5 كتب تقييماً")
root.geometry("900x700")
root.configure(bg="#f9f9f9")

title = tk.Label(root, text="أفضل 5 كتب تقييماً", font=("Arial", 20, "bold"), bg="#f9f9f9", fg="#333")
title.pack(pady=20)

# Function to display each book
def display_book(book):
    frame = tk.Frame(root, bg="white", bd=2, relief="groove", padx=10, pady=10)
    frame.pack(pady=10, fill="x", padx=20)

    # Load and show image
    try:
        response = requests.get(book["thumbnail"])
        img_data = Image.open(BytesIO(response.content))
        img_data = img_data.resize((100, 150))
        img = ImageTk.PhotoImage(img_data)
        img_label = tk.Label(frame, image=img, bg="white")
        img_label.image = img  # keep a reference
        img_label.pack(side="right", padx=10)
    except:
        pass  # If image fails to load, skip it

    # Book info
    info_frame = tk.Frame(frame, bg="white")
    info_frame.pack(side="right", fill="both", expand=True)

    tk.Label(info_frame, text=f"📘 الاسم: {book['Name']}", font=("Arial", 12, "bold"), bg="white", anchor="e").pack(anchor="e")
    tk.Label(info_frame, text=f"✍ المؤلف: {book['Author']}", font=("Arial", 11), bg="white", anchor="e").pack(anchor="e")
    tk.Label(info_frame, text=f"⭐ التقييم: {book['Rate']}", font=("Arial", 11), bg="white", anchor="e").pack(anchor="e")
    tk.Label(info_frame, text=f"📄 عدد الصفحات: {book['Number of pages']}", font=("Arial", 11), bg="white", anchor="e").pack(anchor="e")
    tk.Label(info_frame, text=f"📝 الوصف: {book['Description']}", font=("Arial", 10), bg="white", wraplength=600, justify="right").pack(anchor="e")

# Display all top books
for _, row in top_books.iterrows():
    display_book(row)

root.mainloop()
```
